```python
import pandas as pd
import nltk 
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
data = pd.read_csv("Data_mesh_cleaned.csv")
```

```python
data
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
def analyze_sentiment_nltk(df):
    sentiments = []
    for text in df['pdf_content']:
        sentiment_result = sia.polarity_scores(text)
        
        sentiments.append(sentiment_result)

    df['sentiment'] = sentiments
    return df
```

```python
result_df = analyze_sentiment_nltk(data)
```
